```python
# Helper functions
import numpy as np
import matplotlib.pyplot as plt


def plot_time_signal(signal, sample_rate):
    # Calculate the time vector
    duration = len(signal) / sample_rate
    time_vector = np.linspace(0, duration, num=len(signal))

    # Plot the signal
    plt.figure(figsize=(10, 6))
    plt.plot(time_vector, signal)
    plt.xlabel('Time (seconds)')
    plt.ylabel('Amplitude')
    plt.title('Time Representation of the Signal')
    plt.grid(True)
    plt.show()
```
